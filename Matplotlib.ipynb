### SUBPLOTS

```python
a = np.linspace(0, 10, 11)
b = a ** 4
```

```python
x = np.arange(0, 10)
y = 2 * x
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 2)

axes[0][0].plot(x, y)
axes[1][1].plot(a, b)
axes[0][0].set_title("Inner Title [0][0]")

fig.suptitle("Outer Title")
plt.tight_layout()

fig.savefig('fig.png', bbox_inches='tight')
```

### LEGENDS

```python
x = np.linspace(0, 11, 10)
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x, label = 'X vs X')
ax.plot(x, x**2, label = 'X vs X`2')
ax.legend()
```

### STYLING

```python
x = np.linspace(0, 11, 10)
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x, color = '#E6C017', lw = 2, ls = '--') # RGB HEX CODE
ax.plot(x, x + 1, color = 'purple', linewidth = 3, linestyle = '-.')
```
