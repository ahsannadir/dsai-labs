```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

data = pd.read_csv(r"D:\Datasets\dataset-1.txt")

x = np.array(data["6.1101"]).reshape(-1, 1)
y = np.array(data["17.592"]).reshape(-1, 1)

reg = LinearRegression()
reg.fit(x, y)
y_pred = reg.predict(x)

plt.scatter(x, y)
plt.plot(x, y_pred, color="red")
plt.show()
```
